```python
from transformers import pipeline
```

How Does the question-answering Pipeline Work?
The pipeline("question-answering") automatically loads a pretrained QA model (usually distilbert-base-cased-distilled-squad by default) to extract answers from a given context.

```python
qa_pipeline = pipeline("question-answering")
context = "Hugging Face is a company that focuses on Natural Language Processing."
question = "What does Hugging Face specialize in?"
result = qa_pipeline(question=question, context=context)
print(result)
```

```python
from transformers import pipeline

# Load the Question-Answering model
qa_pipeline = pipeline("question-answering")

def ask_question():
    # Get user input for context and question
    context = input("Enter the context (passage of text): ")
    question = input("Enter your question: ")

    # Get the answer from the model
    result = qa_pipeline(question=question, context=context)

    # Print the extracted answer
    print("\nAnswer:", result["answer"])
    print("Confidence Score:", result["score"])

# Run the function
ask_question()
```
